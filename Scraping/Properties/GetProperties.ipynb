```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
class Pokemon:
    def __init__(self, pokedex_id="", name="", type="", top_image="", abilities=[], hidden_abilities=[], gender_ratio="", evolution=[], generation_introduced="", catch_rate="", egg_group="", height="", weight="", EV_yield="", stats=[], generation_entries=[]):
        self.pokedex_id= pokedex_id
        self.name = name
        self.type = type
        self.top_image = top_image
        self.abilities = abilities
        self.hidden_abilities = hidden_abilities
        self.gender_ratio = gender_ratio
        self.evolution = evolution
        self.generation_introduced = generation_introduced
        self.catch_rate = catch_rate 
        self.egg_group = egg_group
        self.height = height
        self.weight = weight
        self.EV_yield = EV_yield
        self.stats = stats 
        self.generation_entries = generation_entries

    def __str__(self):
        return f"{self.pokedex_id} | {self.name}"
```

```python
with open('Pokemon(wType).json', 'r') as file:
    file_data = json.load(file)

pokemon_data = []

for item in file_data:
    pokemon = Pokemon(
        pokedex_id=item['pokedex_id'],
        name=item['name'],
        type=item['type'],
        top_image=item['top_image'],
        abilities=item['abilities'],
        hidden_abilities=item['hidden_abilities'],
        gender_ratio=item['gender_ratio'],
        evolution=item['evolution'],
        generation_introduced=item['generation_introduced'],
        catch_rate=item['catch_rate'],
        egg_group=item['egg_group'],
        height=item['height'],
        weight=item['weight'],
        EV_yield=item['EV_yield'],
        stats=item['stats'],
        generation_entries=item['generation_entries']
    )
    pokemon_data.append(pokemon)
```

```python
# Insert these names into a base URL to collect the rest of the data
BASE_SEARCH_URL = "https://bulbapedia.bulbagarden.net/wiki/"

def get_attributes(pokemon):
    ability_list = {"Abilities": [], "Hidden Abilities": []}

    response = requests.get(BASE_SEARCH_URL + pokemon.name.replace(" ", "_"))
    soup = BeautifulSoup(response.content, "html.parser")

    titleAbilitie = soup.find('a', title='Ability')
    parent_b = titleAbilitie.find_parent('b')

    try:
        table = parent_b.find_next_sibling('table')

        for td in table.find_all('td'):
            if td.get('style') != 'display: none':
                link = td.find('a')
                if link and link.text != 'Unknown':
                    small_tag = td.find('small')
                    # print(small_tag)
                    # small_text = small_tag.get_text(strip=True)
                    # print(small_text)
                    if small_tag:
                        small_text = small_tag.text.strip()
                        if small_text == pokemon.name:
                            print(f"{link.text} is an Ability")
                            ability_list['Abilities'].append(link.text)
                        elif small_text == "Hidden Ability":
                            print(f"{link.text} is {small_text}")
                            ability_list['Hidden Abilities'].append(link.text)
                    else:
                        print(f"{link.text} is an Ability")
                        ability_list['Abilities'].append(link.text)
    except:
        return

    print(ability_list)
    pokemon.abilities = ability_list['Abilities']
    pokemon.hidden_abilities = ability_list['Hidden Abilities']
```

```python
for pokemon in pokemon_data:
    try:
        print(pokemon)
        get_attributes(pokemon=pokemon)
    except:
        continue
```

```python
import json

def create_json_file(pokemon_list, file_path):
    pokemon_data = []
    for pokemon in pokemon_list:
        pokemon_data.append({
            "pokedex_id": pokemon.pokedex_id,
            "name": pokemon.name,
            "type": pokemon.type,
            "top_image": pokemon.top_image,
            "abilities": pokemon.abilities,
            "hidden_abilities": pokemon.hidden_abilities,
            "gender_ratio": pokemon.gender_ratio,
            "evolution": pokemon.evolution,
            "generation_introduced": pokemon.generation_introduced,
            "catch_rate": pokemon.catch_rate,
            "egg_group": pokemon.egg_group,
            "height": pokemon.height,
            "weight": pokemon.weight,
            "EV_yield": pokemon.EV_yield,
            "stats": pokemon.stats,
            "generation_entries": pokemon.generation_entries
        })

    with open(file_path, 'w') as json_file:
        json.dump(pokemon_data, json_file, indent=4)

create_json_file(pokemon_data, 'Pokemon(wAbilities).json')
```
